## Figures, Axes, Subplot 

```python
plt.figure(figsize=(10,2)) # (figsize=(largura,altura))

plt.plot([10,20,30,40])
```

```python
fig = plt.figure(figsize=(8,6))

ax = fig.add_axes([0.1, 0.1, 1, 0.4])

#Gerando amostra
import numpy as np
nu, sigma = 60, 10
x = nu + sigma * np.random.randn(500)

ax.hist(x, bins=40, facecolor="green", alpha=0.8) # alpha pode ser bom para marcar parte do gráfico
ax.set_xlabel("Score")
ax.set_ylabel("n")
ax.set_title("Histograma", fontsize=18)
ax.spines["right"].set_visible(False) # retira linha da direita
ax.spines["top"].set_visible(False) # retira linha de cima
ax.grid(True)
```

```python
plt.hist(x, bins=50, facecolor="green", alpha=0.8);
```

```python
fig = plt.figure(figsize=(8,6))

ax = fig.add_axes([0.1, 0.1, 1, 0.4])

import numpy as np
nu, sigma = 60, 10
x = nu + sigma * np.random.randn(500)

ax.hist(x[x>=50], bins=50, facecolor="red", alpha=0.99);
ax.hist(x[x<50], bins=50, facecolor="red", alpha=0.35);
```

## Outros gráficos

```python
plt.bar(x=[1,2,3,4],height=[1,4,9,16], width=.75); # precisa de dois parametros
```

```python
# vamos alterar o tamanho
fig = plt.figure(figsize=(8,6));

# é possível mudar algumas coisas inclusive parametros nao obrigatorios
plt.bar(x=[1,2,3,4],
        height=[1,4,9,16],
        color="r",
        width=.8,
        edgecolor="Black",
        tick_label=["Grupo A", "Grupo B", "Grupo C", "Grupo D"],
        bottom=5
        # log=True
        );

# e incrementar com o que já foi visto
plt.xlabel("Eixo X", fontsize=12);
plt.ylabel("Eixo Y", fontsize=12);
plt.title("Graf de Barras", fontsize=18, color="Black");
```

```python
plt.barh([1,2,3,4],[1,4,9,16]);
plt.title("Graf. Horizontal")
```

```python
graf_bar = plt.bar([1,2,3,4],[1,4,9,16], tick_label=["Grupo A", "Grupo B", "Grupo C", "Grupo D"]);
graf_bar[0].set_color("red")
graf_bar[1].set_color("green")
graf_bar[2].set_color("orange")
```

```python

import pandas as pd

housing = pd.read_csv("data\Housing.csv")
housing.head()
```

```python
housing_median = plt.bar(x=housing.groupby("furnishingstatus")["bedrooms"].count().index, 
                         height=housing.groupby("furnishingstatus")["bedrooms"].count())

housing_median[0].set_color("red")
housing_median[1].set_color("orange")
housing_median[2].set_color("yellow")

plt.grid(True)
```

```python
import seaborn as sns

sns.catplot(x="furnishingstatus", kind="count", data=housing);
```

```python
sns.regplot(x="area", y="price", data=housing); # grafico de dispersao com regressao linear
```
